```python
#Imprt the lybrary
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(f'C:\\Users\\taufi\\OneDrive\\Desktop\\coding wala folder\\Netflix_Data_Visualisation\Project\\netflixcsv.csv')
# df.to_excel("output.xlsx", index=False)

# print(df.head())

#clean vlau
df = df.dropna(subset=['type','release_year','rating','country','duration'])
type_count = df['type'].value_counts()
plt.figure(figsize=(6,4))
plt.bar(type_count.index , type_count.values, color=['blue','red'])

# Add title

plt.title("Number of Movies VS TV Shows on Netflix")
plt.xlabel('Type')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig("Movie_Vs_tvshows")
plt.grid()
plt.show()

#Creat pie Chart
rating_count = df['rating'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(rating_count,labels=rating_count.index,autopct= '%1.1f%%',startangle=90)
plt.title("Percentege og Content rating")
plt.tight_layout()
plt.savefig('Conten_Rating')
plt.show()


movie_df = df[df['type']=='Movie'].copy()
movie_df['duration_int'] = movie_df['duration'].str.extract('(\d+)').astype(float)

plt.figure(figsize=(8,6))
plt.hist(movie_df['duration_int'],bins=30,color='purple',edgecolor='green')
plt.title("Distrubtion of movie Duration")
plt.xlabel('Duration (minut)')
plt.ylabel('Number of Movies')
plt.tight_layout()
plt.savefig("Movie_duration_histo.png")
plt.grid()

plt.show()

#Scater plot
release_counts = df['release_year'].value_counts().sort_index()
plt.figure(figsize=(10,6))
plt.scatter(release_counts.index,release_counts.values,color='red')
plt.title("Relese year Vs Nof of shows")
plt.xlabel('Relese year')
plt.ylabel('Number of shows')
plt.tight_layout()
plt.savefig("Relese_Year_Vs_nos.png")
plt.grid(color='black')
plt.show()

#No of Counry Who Most creat movie
country_count = df['country'].value_counts().head(10)
plt.figure(figsize=(8,6))
plt.barh(country_count.index,country_count.values,color='teal')
plt.title("Top 10 Contery by Nuber of shows")
plt.xlabel('Number of Shows')
plt.ylabel('Country')
plt.tight_layout()
plt.savefig("Top10-country.png")

content_by_year = df.groupby(['release_year','type']).size().unstack().fillna(0)

fig, ax = plt.subplots(1, 2, figsize=(12,5))

#First subplot:movies
ax[0].plot(content_by_year.index , content_by_year['Movie'],color='blue')
ax[0].set_title('Movies Relesed Per Year')
ax[0].set_xlabel("year")
ax[0].set_ylabel('No of moves')

#Creat send Subplot"tv SHow
ax[0].plot(content_by_year.index, content_by_year['TV Show'], color='orange')
ax[0].set_title('Tv Shows Relesed Per Year')
ax[0].set_xlabel("year")
ax[0].set_ylabel('No of moves')

fig.suptitle('Comparision of movies and TV Shows relese Years')

plt.tight_layout()
plt.savefig('movies_TVShow_comp')
```
